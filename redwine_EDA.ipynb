```python
# Categorical Plot
sns.catplot(data = df , x = 'quality' , y = 'alcohol', kind = 'box' )
# feature alcohol is plotted for every quality of wine 
# Conclusion only one outlier in Quality '3' & '6', more  number of outliers are in quality '5', 
# no outliers in quality '4','7' & '8'
```
